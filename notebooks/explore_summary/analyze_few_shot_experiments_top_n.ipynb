```python
import sys
sys.path.insert(0,'/workspaces/RAG_secure_code_generation/src')
sys.path.insert(0,'/workspaces/RAG_secure_code_generation')
import pandas as pd
import os
```

```python
summary_file = "../../experiments/task_detect_xss_simple_prompt/experiments_summary.csv"
output = "../../output/few_shot_experiments_{examples}_examples_top_{n}.csv"
```

```python
parameters_to_keep = [
    "model_name",
    "temperature",
    "successes",
    "examples_per_class"  
]
top_n_parameters_to_keep = [
    "accuracy",
    "precision",
    "recall",
    "f1"
]

top_n = 10
examples_per_class = 20
```

```python
seed = 156
```

```python
output = output.format(examples=examples_per_class, n = top_n)
output
```

```python
parameters_to_keep.extend(list(map(lambda x: f"top_{top_n}_{x}",top_n_parameters_to_keep)))
parameters_to_keep
```

```python
df = pd.read_csv(summary_file)
df.head()
```

```python
#keep only the rows with seed
df = df[df["seed"] == seed]
df.head()
```

```python
#filter keeping only generation mode few_shot
df = df[df["generation_mode"] == "few_shot"]
df.head()
```

```python
#filtering keeping only examples_per_class
df = df[df["examples_per_class"] == examples_per_class]
df.head()
```

```python
#keep only the parameters we want
df = df[parameters_to_keep]
df.head()
```

```python
#set the precision of the metrics to 3f
df = df.round(3)
df.head()
```

```python
#drop examples_per_class column
df = df.drop(columns=["examples_per_class"])
df.head()
```

```python
#sort by model name and by temperature
df = df.sort_values(by=["model_name", "temperature"])
df.head()
```

```python
len(df)
```

```python
#save the dataframe
df.to_csv(output, index=False)
```
